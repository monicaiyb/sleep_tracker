## Dealing with missing data technique 

### 1: Dropping missing values
One way to deal with missing data is to drop it. This is particularly useful in cases where you have plenty of data and losing a small portion won’t impact the downstream analysis.

```python
df3 = df2.copy()
df3 = df3.dropna()
df3.shape
```

```python
# The default axis removes the rows containing NaNs. Use axis = 1 to remove the columns with one or more NaN values.
df3 = df2.copy()
df3.dropna(inplace=True, axis=1)
df3.head()
```

### 2: Replacing missing values
Instead of dropping, replacing missing values with a summary statistic or a specific value (depending on the use case) maybe the best way to go.
df3 = df2.copy()
# Get the mean of Pregnancies
mean_value = df3['Age'].mean()
# Fill missing values using .fillna()
df3 = df3.fillna(mean_value)

# Visualization with Python
Data visualization transforms information into a graphical medium. It uses different charts to clarify complex datasets, making them easier to understand.

Graphs, charts, and maps represent the various types of visualizations. Each puts the data sets into an intuitive graphical context to help reveal patterns and correlations.

## Matplotlib
Matplotlib is a low-level, user-friendly library that offers flexibility. It works with various dataset fields and arrays to plot basic graphs and charts. This nifty tool provides a range of useful features, like:

3D plots
Line graphs
Scatter charts
Histograms
Bar charts

```python
fig = plt.figure(figsize=(8, 6))
plt.show()
```

This is the overall layout space, a container where one or more plots (called Axes) will be placed. Alone, it’s just a blank canvas.

```python
# adding Axes manually to the figure
# Axes is the actual plot area — the box where data gets drawn.
fig = plt.figure(figsize=(8, 6))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # [left, bottom, width, height] in 0–1 figure coords
plt.show()
```

```python
# common shortcut — creates both Figure and Axes together
fig, ax = plt.subplots(figsize=(8, 6))
plt.show()
```

```python
# adding a label to the plot line — needed for legends
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(sleep_df["Age"], sleep_df["Occupation"], label="Occupation")
ax.set_xlabel("Age", fontweight = 'semibold', fontsize='9')
ax.set_ylabel("Quality of Sleep ", fontweight = 'semibold', fontsize='9')
ax.set_title("Age vs Occupation", fontweight='bold', fontsize=20)
plt.grid(True)
plt.show()
```

```python
# Scatter plot with Occupation against Age
plt.scatter(sleep_df['Occupation'], sleep_df['Age'])
```

```python
# Scatter plot with Occupation against Age
fig, ax = plt.subplots(figsize=(8, 6))
plt.scatter(sleep_df['Occupation'], sleep_df['Age'], s=50, facecolor='C2', edgecolor='k')
# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Occupation')
plt.ylabel('Age')
plt.title('Occupation Vs Age', fontsize = 25)
plt.xticks(fontsize = 6)
plt.yticks(fontsize = 10)

plt.show()
```
