## DataFrame

A DataFrame is a table. It contains an array of individual entries, each of which has a certain value. Each entry corresponds to a row (or record) and a column.

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df = pd.DataFrame(data)

print(df) 
```

## I want to store passenger data of the Titanic. 
For a number of passengers, I know the name (characters), age (integers) and sex (male/female) data.

```python
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss Elizabeth",
        ],

        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],

    }

)
df
```

The list of row labels used in a DataFrame is known as an Index. 
We can assign values to it by using an index parameter in our constructor:

```python
df=pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
df
```

## Reading data files

Data can be stored in any of a number of different forms and formats. By far the most basic of these is the humble CSV file. When you open a CSV file you get something that looks like this:

Product A,Product B,Product C,
30,21,9,
35,34,1,
